```python
import json

def load_json_file(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    return data
def write_to_json_file(data, file_path):
    with open(file_path, 'w') as file:
        json.dump(data, file, indent=4)
```

### prepare paths and details of these 5 videos (from low 10 decap cider) TO RUN ZeroCap on them

```python
import os
def find_file_path(directory, filename):
    for root, dirs, files in os.walk(directory):
        if filename in files:
            return os.path.join(root, filename)

    return None
val1_caps = load_json_file("./an_cap_paragraph_summary_val_1.json")


val_dir_path = "/home/grads/h/hasnat.md.abdullah/open_ended_activity_analysis/zero-shot-video-to-text/data/ActivityNet_captions_dataset/validation"

video_ids = [
    "v_4Lu8ECLHvK4",
    "v_yE5whKJ-DE4",
    "v_sRN_crwj3B4",
    "v_OPp3DqFq0O0",
    "v_1926p23ooUM"
]
for id in video_ids:
    for k,v in val1_caps.items():
        if id == k:

            # print(k)
            # print(v)
            video_id = k[2:]
            file_name = video_id+".mp4"
            file_path = find_file_path(val_dir_path,file_name)

            v['path'] = file_path
            val1_caps[k] = v

            break

experiment_decap_low10_videos = {}
for id in video_ids:
    for k,v in val1_caps.items():
        
        if k == id:
            experiment_decap_low10_videos[k]=v
            

experiment_decap_low10_videos
write_to_json_file(experiment_decap_low10_videos,"./experiment_decap_low10_videos.json")
```
