```python
import json
def load_json_file(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    return data
```

```python
# val-va1 
data = load_json_file("/home/grads/h/hasnat.md.abdullah/open_ended_activity_analysis/zero-shot-video-to-text/experiments/summary_study_activityNet/decap_zero_cap_output_files/decap_output_val_1_summary_id_caption_tian.json")
# Sort the data based on DeCap_caption_cider in ascending order
sorted_data = sorted(data.items(), key=lambda x: x[1]["DeCap_caption_cider"])

# Get the summary_cap, generated_summary, and DeCap_caption with the highest 10 DeCap_caption_cider values
highest_10 = sorted_data[-10:]
highest_10 = [(d[1]["summary_cap"], d[1]["generated_summary"], d[1]["DeCap_caption"]) for d in highest_10]

# Get the summary_cap, generated_summary, and DeCap_caption with the lowest 10 DeCap_caption_cider values
lowest_10 = sorted_data[:10]
lowest_10 = [(d[1]["summary_cap"], d[1]["generated_summary"], d[1]["DeCap_caption"]) for d in lowest_10]

# Print the results
print("Highest 10 DeCap_caption_cider values:")
for i, item in enumerate(highest_10, 1):
    print(f"{i}.")
    print("   summary_cap:", item[0])
    print("   generated_summary:", item[1])
    print("   DeCap_caption:", item[2])
    print()

print("Lowest 10 DeCap_caption_cider values:")
for i, item in enumerate(lowest_10, 1):
    print(f"{i}.")
    print("   summary_cap:", item[0])
    print("   ZeroCap_caption:", item[1])
    print("   DeCap_caption:", item[2])
```
