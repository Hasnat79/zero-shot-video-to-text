### train set: How many videos have summary captions?

```python
sum =0
for id in train_caps.keys():
    time_stamps = train_caps[id]['timestamps']
    if time_stamps[0][-1] == time_stamps[-1][-1]:
        total_video_segment = time_stamps[0]
        # print(total_video_segment)
        sum+=1
print(f"Total videos with summary captions: {sum}")
```

```python
sum =0
train_summary_id_to_caption ={}
for id in train_caps.keys():
    time_stamps = train_caps[id]['timestamps']
    # end time of first segment matches with end time of last segment
    if time_stamps[0][-1] == time_stamps[-1][-1]:
        total_video_segment = time_stamps[0]
        # print(id)
        # print(train_caps[id])
        start = time_stamps[0][0]
        end = time_stamps[0][1]
        cap = train_caps[id]['sentences'][0]
        # print(f"""
        #       start: {start}
        #       end: {end}
        #       summary_cap = {cap}
        #       """)
        train_summary_id_to_caption[id]={
            "start":start,
            "end":end,
            "summary_cap":cap,
            "path":""
        }
      
len(train_summary_id_to_caption)
```
