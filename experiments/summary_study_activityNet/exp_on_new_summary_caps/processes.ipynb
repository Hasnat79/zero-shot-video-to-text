### merging tian and hasnat's captions on low10 videos into one combined_manual_cap file

```python
import json

def load_json_file(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    return data
def write_to_json_file(data, file_path):
    with open(file_path, 'w') as file:
        json.dump(data, file, indent=4)


hasnat_low10_path ="/home/grads/h/hasnat.md.abdullah/open_ended_activity_analysis/zero-shot-video-to-text/experiments/summary_study_activityNet/exp_on_new_summary_caps/decap_output_val_1_summary_id_caption_hasnat_low10 copy.json"
tian_low10_path = "/home/grads/h/hasnat.md.abdullah/open_ended_activity_analysis/zero-shot-video-to-text/experiments/summary_study_activityNet/exp_on_new_summary_caps/decap_output_val_1_summary_id_caption_tian_low10 copy.json"

hasnat_low10 = load_json_file(hasnat_low10_path)
tian_low10 = load_json_file(tian_low10_path)

#grab Tian's captions in a dict
tian_caps = {}
for i in tian_low10:
    video_id = i[0]
    video_details = i[1]

    tian_caps[video_id] = video_details["tian_caption_no_instruction"]

# put  tian_caps on hasnat's caption file
for i in hasnat_low10:
    video_id = i[0]
    video_details = i[1]
    
    video_details["tian_caption_no_instruction"] = tian_caps[video_id]

# hasnat_low10
write_to_json_file(hasnat_low10,"./decap_output_val_1_summary_id_caption_combined_manual_cap_low10.json")
```
