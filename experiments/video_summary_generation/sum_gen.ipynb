## selecting the first video of each label (50 videos)

```python

def extract_title(url):
    import requests
    from bs4 import BeautifulSoup


    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        title_tag = soup.find('title')
        if title_tag:
            return title_tag.string
    return None
    

    return title
```

```python
video_details = []


for label in least_50_label_video_id_segment_caps.keys():
    # print(label)
    data = {}
    data['label'] = label
    
    # print("len(least_50_label_video_id_segment_caps[label].keys()):", len(least_50_label_video_id_segment_caps[label].keys()))
    for video_id in least_50_label_video_id_segment_caps[label].keys():
        
        video = least_50_label_video_id_segment_caps[label][video_id]
        data["video_id"] = video_id
        # print("id:", video_id)
        
        # print(video[0])
        data['path'] = video[0]['path']
        yt_link = f"https://www.youtube.com/watch?v={video_id}"
        data['yt_link'] = yt_link
        # print(video_id)
        title = extract_title(yt_link)
        if title == " - YouTube":
            continue
        data['title'] = title
        video_details.append(data)
        break


    # break
video_details
```
